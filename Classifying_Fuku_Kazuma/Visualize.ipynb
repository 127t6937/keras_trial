```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline

script_path = os.path.dirname("./")
print(script_path)

k = 1

for f in os.listdir("./Fuku_merge"):
    image = Image.open("./Fuku_merge/" + f)
    if k > 5:
        continue

    plt.tick_params(labelbottom='off')
    plt.tick_params(labelleft='off')
    plt.axis('off')
    plt.subplot(1, 5, k)
    plt.imshow(image)
    k = k + 1

plt.tick_params(labelbottom='off')
plt.tick_params(labelleft='off')
plt.axis('off')
#plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
